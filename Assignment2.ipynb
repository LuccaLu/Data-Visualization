```python
import pandas as pd
import matplotlib.pyplot as plt

sleep_data = pd.read_csv('sleep.csv')

bar_colors = ['#0c3759', '#4789be']

plt.subplots(figsize=(10, 6))

index = sleep_data.index
bar_width = 0.35
gap = 0.05

plt.barh(index + bar_width/2 + gap, sleep_data['Recommended lower'], bar_width, label='Recommended lower limit', color=bar_colors[1], edgecolor='none')
plt.barh(index - bar_width/2 - gap, sleep_data['Recommended upper'], bar_width, label='Recommended upper limit', color=bar_colors[0], edgecolor='none')


plt.text(0.5, 1.08, "National Sleep Foundation's recommended hours of sleep",
         horizontalalignment='center',
         fontsize=16,
         transform=plt.gca().transAxes)

plt.text(0.3, 1.02, "Recommended lower limit",
         horizontalalignment='center',
         fontsize=12,
         color=bar_colors[1],
         transform=plt.gca().transAxes)

plt.text(0.8, 1.02, "Recommended upper limit",
         horizontalalignment='center',
         fontsize=12,
         color=bar_colors[0],
         transform=plt.gca().transAxes)

plt.xlabel('Hours', fontsize=12)
plt.yticks(index,sleep_data['Age group'])
plt.gca().invert_yaxis()
plt.xlim(0, 18)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

dot_colors = ['#0c3759', '#4789be']

sleep_data_inverted = sleep_data.iloc[::-1].reset_index()

for i, row in sleep_data_inverted.iterrows():
    ax.plot([row['Recommended lower'], row['Recommended upper']], [i, i], '-', color='#000000')
    ax.plot(row['Recommended lower'], i, 'o', color=dot_colors[0], markersize=8, markeredgecolor='black')
    ax.plot(row['Recommended upper'], i, 'o', color=dot_colors[1], markersize=8, markeredgecolor='black')

ax.set_yticks(range(len(sleep_data_inverted)))
ax.set_yticklabels(sleep_data_inverted['Age group'])
ax.tick_params(axis='y', which='both', left=False)

ax.xaxis.tick_top()
ax.set_xticks([0, 6, 8, 10, 12, 14, 18, 24])
ax.set_xlim(0, 24)

ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

ax.set_title("National Sleep Foundation's recommended hours of sleep", pad=20, fontsize=16)

plt.show()
```
